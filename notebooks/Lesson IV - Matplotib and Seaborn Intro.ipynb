## 2. Figures y Axes

Las dos estructuras fundamentales en `matplotlib` son las `Figures` y los `Axes`.

- **Figure**: Es el contenedor principal de la gr치fica o gr치ficas. Puede contener uno o varios `Axes`. Imaginalo como un canvas en el que vas a pintar tu gr치fica.
- **Axes**: Cada una de las gr치ficas que componen el canvas. Es donde "se dibujan" los datos. Una `Figure` puede tener varios `Axes`.

![figures](https://miro.medium.com/v2/resize:fit:2000/1*CANkzqegZcHv9znE1U4s7A.png)

### 2.1 Crear una Figure

Para crear una `Figure`, simplemente utilizamos la funci칩n `plt.figure()` de `matplotlib`.

````python
fig = plt.figure(figsize=(10, 6)) 
````

### 2.2 Crear un Axes

Para crear un `Axes`, utilizamos la funci칩n `fig.add_subplot()`. Esta funci칩n recibe 3 par치metros:

- **nrows**: N칰mero de filas de la rejilla de `Axes`.
- **ncols**: N칰mero de columnas de la grilla de `Axes`.
- **index**: 칈ndice del `Axes` que queremos crear.

游녢 Este c칩digo se lee de la siguiente forma: 

Si divides la figura en una **rejilla de 2 filas y 3 columnas**, el siguiente **gr치fico** que se a침ada **se ubicar치 en la primera posici칩n**. Es decir, en la **esquina superior izquierda.** La numeraci칩n de las posiciones de la rejilla es de izquierda a derecha
y de arriba a abajo.

````python
fig = plt.figure()  # en cada celda necesitamos una figura nueva
ax1 = fig.add_subplot(2, 3, 1)
ax2 = fig.add_subplot(2, 3, 6)

# 游눠 importante, matplotlib autom치ticamente no muestra los ejes no creados
````

En `matplotlib` existen (por desgracia 游땐) m칰ltiples formas de crear `figures` y `axes`, incluso al mismo tiempo, algunas de las otras formas por las que puedes crear una `figura` y uno o varios `axes` son las siguientes:

### 4.1 Gr치ficas de l칤nea

Las gr치ficas de l칤nea son muy 칰tiles para visualizar la **evoluci칩n de una variable a lo largo del tiempo**. Por ejemplo, si tienes datos de ventas de una empresa, puedes utilizar una gr치fica de l칤nea para ver c칩mo han evolucionado las ventas a lo largo de los a침os.

Para crear una gr치fica de l칤nea con `seaborn`, utilizamos la funci칩n `sns.lineplot()`. Esta funci칩n recibe los siguientes par치metros:

- **x**: Valores del eje x: puede ser el nombre de una columna de un DataFrame o una lista de valores.
- **y**: Valores del eje y: puede ser el nombre d euna columna de un DataFrame o una lista de valores.
- **data**: DataFrame que contiene los datos.

````python
sns.lineplot(x="time", y="total_bill", data=tips_dataset)
````

游눠 `seaborn` gestiona la creaci칩n del `ax` por debajo, simplificando el c칩digo, aun as칤 es posible pasarle un `ax` personalizado por par치metro.

````python
fig = plt.figure(figsize=(10, 3))

ax1 = fig.add_subplot(1, 3, 1)
ax2 = fig.add_subplot(1, 3, 3)

sns.lineplot(x="time", y="total_bill", data=tips_dataset, ax=ax2)
````

Hay un cuarto par치metro, `hue`, que nos permite a침adir una tercera dimensi칩n a la gr치fica. Por ejemplo, si queremos ver c칩mo ha evolucionado la propina en funci칩n del d칤a de la semana, podemos utilizar `hue="day"`.
